```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('assets/data/enfermedades.csv')
datos.head(5)
```

```python
datos['Categoria'] = datos['Categoria'].fillna(0)
datos
```

```python
hombres_respiratoria = datos[datos['Categoria'].apply(lambda x: 'Respiratoria' in x)]['Hombres']
promedio_h_respiratoria = hombres_respiratoria.mean()
promedio_h_respiratoria
```

```python
mujeres_respiratoria  = datos[datos['Categoria'].apply(lambda x: 'Respiratoria' in x)]['Mujeres']
promedio_m_respiratoria = mujeres_respiratoria.mean()
promedio_m_respiratoria
```

```python
datos_promedio = pd.DataFrame({'Categoria': ['Hombres', 'Mujeres'], 'Promedio' : [promedio_h_respiratoria,promedio_m_respiratoria]})
```

```python
plt.pie(datos_promedio['Promedio'],labels = datos_promedio['Categoria'], autopct = '%1.1f%%')
plt.title('Enfermedades Respiratorias: Hombres y Mujeres')
plt.show()
```

```python
def respiratorias(Categoria):
    return 'Neurológica' in Categoria

datos_respiratorio = datos[datos['Categoria'].apply(respiratorias)]
enfermedades = datos_respiratorio[['Enfermedad','Hombres']]
enfermedades.head(50)
```

```python
Dataset_Respiratorias = datos_respiratorio.head(10)
Dataset_Respiratorias.set_index('Enfermedad', inplace=True)

ax = Dataset_Respiratorias['Hombres'].plot(x=None, kind='bar')
for i, enfermedad in enumerate(Dataset_Respiratorias['Hombres']):
    ax.annotate(str(enfermedad), xy=(i, enfermedad), ha='center', va='bottom')
```

```python
def respiratorias_m(Categoria):
    return 'Neurológica' in Categoria

datos_respiratorio_m = datos[datos['Categoria'].apply(respiratorias_m)]
datos_m = datos_respiratorio_m[['Enfermedad', 'Mujeres']]
datos_m.head(10)
```

```python
Dataset_Respiratorias_m = datos_m.head(10)
Dataset_Respiratorias_m.set_index('Enfermedad', inplace=True)
ax = Dataset_Respiratorias_m['Mujeres'].plot(x=None, kind='bar')
for i, enfermedad in enumerate(Dataset_Respiratorias_m['Mujeres']):
    ax.annotate(str(enfermedad),xy= (i,enfermedad), ha='center', va='bottom')
```
